```python
import pandas as pd

def convert_categorical_columns(df, cat_columns):
   
    label_mappings = {}

    for col in cat_columns:
       
        unique_labels = df[col].unique()
        
        label_dict = {label: num for num, label in enumerate(unique_labels)}
      
        label_mappings[col] = label_dict
        
        df[col] = df[col].map(label_dict)
    
    return df, label_mappings
```

```python
cat_columns = ['case_id', 'continent']
df, mappings = convert_categorical_columns(df, cat_columns)
```
