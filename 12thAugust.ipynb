```python
import pandas as pd

# Example DataFrame
df = pd.DataFrame({
    'age': [25, 30, 35],
    'job': ['admin', 'blue-collar', 'technician'],
    'salary': [50000, 60000, 70000]
})

# Get the data types of each column
types = df.dtypes

# List of categorical columns (object type)
cat_columns = [key for key, value in types.items() if value == 'object']

# List of numerical columns (excluding object type)
num_columns = [key for key, value in types.items() if value != 'object']

print("Categorical columns:", cat_columns)
print("Numerical columns:", num_columns)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Example DataFrame
continents_df = pd.DataFrame({
    'continent': ['Asia', 'Europe', 'Africa', 'America'],
    'number_of_applicants': [150, 100, 80, 70]
})

# Compute the value counts of the 'continent' column
counts = continents_df['number_of_applicants']

# Create a bar chart
plt.figure(figsize=(8, 5))
plt.bar(continents_df['continent'], counts)

# Set title and labels
plt.title("Bar Chart of Number of Applicants by Continent")
plt.xlabel('Continent')
plt.ylabel('Number of Applicants')

# Save the plot
plt.savefig('continents_bar_chart.jpg')

# Display the plot
plt.show()
```
